```python
# Categorizing Ratings

unpleasant = df[df['Rating']  < 2.0]
dissappointing = df[(df['Rating'] < 3.0) & (df['Rating'] > 1.0)]
satisfactory = df[(df['Rating'] >= 3.0) & (df['Rating'] <= 3.75)]
Premium  = df[(df['Rating'] <= 4.0) & (df['Rating'] >= 3.76)]
Elite = df[(df['Rating'] <= 5.0) & (df['Rating'] > 4.0)]
```

```python
# Finding the rating distribution count

Ratings = [unpleasant.shape[0],dissappointing.shape[0],satisfactory.shape[0],Premium.shape[0],Elite.shape[0]]
Ratings
```

```python
# A plot of the rate distribution

fig, ax = plt.subplots(figsize=[16,4])
sns.distplot(df['Rating'], ax=ax)
ax.set_title('Rating, Distribution')
plt.savefig('Rating_Distribution.png')
plt.show()
```

```python
#Counutries with highest contribution to rating

count_rating = pd.concat([pd.Series(row['Rating'], row['Company Location'].split(',')) for _, row in df.iterrows()]
         ).reset_index()
count_rating.columns = ['Company Location', 'Rating']
count_rating['mean_rating'] = count_rating.groupby(['Company Location'])['Rating'].transform('mean')

## Look at boxplot over the countries (contributors in blends)
fig, ax = plt.subplots(figsize=[6, 16])
sns.boxplot(
    data=count_rating.sort_values('mean_rating', ascending=False),
    y='Company Location',
    x='Rating'
)
ax.set_title('Boxplot, Rating for countries (contributors)')
plt.savefig('General Country Rating.png')
plt.show()
```

```python
# Top 5 Countries with highest ratings

countries=df['Company Location'].value_counts().index.tolist()[:5]
best_choc={} # empty dictionary
for j in countries:
    c=0
    b=df[df['Company Location']==j]
    br=b[b['Rating']>=3] # rating more than 3
    for i in br['Rating']:
        c+=1
        if(j=='U.S.A.'):
            j='USA'
        best_choc[j]=c    
# Code to visualize the countries that give best cocoa beans
li=best_choc.keys() # li is list of the keys in the best_choc dictionaries

# The barplot is used
plt.bar(range(len(best_choc)), best_choc.values(), align='center'
        ,color=['#00811F','#FFC0CB','#faacff','#d04949','#e59a9a'])
plt.xticks(range(len(best_choc)), list(li))
plt.xlabel('Company Location',color='brown',size=15)
plt.ylabel('Count of Ratings above 3',color='brown',size=15)
plt.savefig('Company_Location.png')
plt.show()
```

```python
## Finding country chocolate ratings over the years
## using the average rating

flow = pd.crosstab(
    df['Company Location'],
    df['Review Date'],
    df['Rating'], aggfunc='mean'
)
flow['tot'] = flow.sum(axis=1)
flow = flow.sort_values('tot', ascending=False)
flow = flow.drop('tot', axis=1)

fig, ax = plt.subplots(figsize=[10,6])
sns.heatmap(flow.head(20), cmap='cubehelix', linewidths=.5)
ax.set_title('Company location, mean rating over the years')
plt.savefig('Rate_changes_over_years.png')
plt.show()
```

```python
#finding correlation between cocoa percentage and ratings
sns.regplot(x='Cocoa Percentage(%)',y='Rating',data=df)
plt.ylim(0,)
plt.savefig('correlation_percentage_rating.png')
#shows there is a weak negative correlation between the percentage of cocoa in the chocolate and the rating 


## Pearson correlation
pearson_coef,p_value=stats.pearsonr(df['Cocoa Percentage(%)'],df['Rating'])
print("Pearson Coefficient:",pearson_coef)   #value shows us there is indeed a weak negative correlation
print('Pvalue:',p_value)     #value shows us we are very certain about the correlation we calculated


#a negative correlation means that as one variable increases the other decreases,
#meaning as the percentage of cocoa in the chocolate increases the rating decreases
#a weak correlation means that as the percentage of cocoa in the chocolate increases the ratings decrease but only slightly, by 0.16  


## Finding the unique values in the percentage column
percentages=df['Cocoa Percentage(%)'].unique()
print("\n\nThe different percentages:",percentages)


## Finding the minimum percentage value
minimum=df['Cocoa Percentage(%)'].min()
print("\n\n Minimum percentage:",minimum)


## Finding the unique values in the review year
years=df['Review Date'].unique()
print("\n\nThe different years:",years)


## Binning the values in the percentage column
bins=np.linspace(min(df['Cocoa Percentage(%)']),max(df['Cocoa Percentage(%)']),5)
print('\n',bins)
```
